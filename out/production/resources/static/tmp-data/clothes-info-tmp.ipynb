```python
import pandas as pd
```

```python
# 카테고리
category = ['상의', '하의', '아우터', '원피스']
top = ['탑', '블라우스', '티셔츠', '니트웨어', '셔츠', '브라탑', '후드티']
bottom = ['청바지', '팬츠', '스커트', '레깅스', '조거팬츠']
outer = ['코트', '재킷', '점퍼', '패딩', '베스트', '가디건', '집업']
dress = ['드레스', '점프수트']
```

```python
# 기장
top_length = ['크롭', '노멀', '롱']
bottom_length = ['미니', '니렝스', '미디', '발목', '맥시']
outer_length = ['크롭', '노멀', '하프', '롱', '맥시']
dress_length = ['미니', '니렝스', '미디', '발목', '맥시']
```

```python
# 핏
bottom_fit = ['스키니', '노멀', '와이드', '루즈', '벨보텀']
fit = ['타이트', '노멀', '루즈', '오버사이즈']
```

```python
# 소재
material = ['퍼', '니트','무스탕'  '레이스', '스웨이드', '린넨', '앙고라' , '메시', '코듀로이' , '플리스', '시퀸/글리터' ,'네오프렌'
'데님', '실크' ,'저지'  ,'스판덱스', '트위드'  '자카드', '벨벳'  '가죽', '비닐/PVC' , '면' ,'울/캐시미어' , '시폰', '합성섬유']
```

```python
# 색상
color = ['블랙','화이트','그레이','레드','핑크','오렌지','베이지','브라운','옐로우','그린',
         '카키','민트','블루','네이비','스카이블루','퍼플','라벤더','와인','네온','골드']
```

# csv 파일 만들기

```python
clothes_info = pd.DataFrame()
clothes_info['category_l'] = None
clothes_info['category_s'] = None
clothes_info['length'] = None
clothes_info['fit'] = None
clothes_info
```

```python
clothes = [top, bottom, outer, dress]
length = [top_length, bottom_length, outer_length, dress_length]
fits = [fit, bottom_fit, fit, fit]
```

```python
for large, small, l, f in zip(category, clothes, length, fits):
  
  for small_entry in small: 
    for length_entry in l:
      for f_entry in f:
        location = clothes_info.shape[0]
        clothes_info.loc[location] = [large, small_entry, length_entry, f_entry]
```

```python
clothes_info = clothes_info.reset_index()
```

```python
clothes_info = clothes_info.rename(columns={'index':'id'})
```

```python
clothes_info
```

```python
clothes_info.to_csv('clothes_info.csv', index=False, encoding='euc-kr')
```
